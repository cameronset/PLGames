```python
import requests
import pandas as pd
import time
import numpy as np
#driver = webdriver.Chrome()

#url = "https://www.whoscored.com/Regions/252/Tournaments/2/England-Premier-League"

file = open("Teams.txt")
tems = file.read().split(",")

#url = "https://www.soccerstats.com/results.asp?league=england"
url = "https://www.soccerstats.com/results.asp?league=england&pmtype=bydate"


respe = requests.get(url)
tems

teams = [x[1:] for x in tems]
teams[0] = "Arsenal"
teams
```

```python
import bs4

soup = bs4.BeautifulSoup(respe.text, 'html.parser')
soup
```

```python
#tr_elements = soup.xpath('//tr')
#tr_elements
tag = soup.find_all('table', {'cellpadding':'1', 'cellspacing':'0', 'style':'margin-left:10px;'})
tag
```

```python
#tagg = tag.find('td', {'align':'right'})
home = soup.find_all('td', {'align':'right'})
away = soup.find_all('td', {'align':'left'})
score = soup.find_all('td', {'align':'center', "style":"border:1px solid #f0f0f0;"})
score

xx=[t.text for t in home]
yy = [t.text for t in away]
score = [t.text for t in score]
xx
yy
score
```

```python
#score = [x for x in score if ":" not in x]




homeGoals = [x[:1] for x in score]



score
```

```python
homeTeams = [x for x in xx if "\xa0"  in x]
homeTeams
homeTeams = [x[:-1] for x in homeTeams]
homeTeams


#for x in homeTeams:
#    if x not in teams:
#        homeTeams.remove(x)
#homeTeams


homeTeams = [p for p in homeTeams if p in teams]
homeTeams
```

```python
awayTeams = [x for x in yy if "\xa0" in x]
awayTeams
awayTeams = [x[1:] for x in awayTeams]
awayTeams

awayTeams = [p for p in awayTeams if p in teams]
awayTeams
```

```python
df = pd.DataFrame({'Home Team':homeTeams[:len(score)], "Score":score, "Away Team":awayTeams[:len(score)]})
pd.set_option('display.max_rows', None)
df
```

```python
df = df[df["Score"] != "pp."]
df = df[~df.Score.str.contains(":")]
df
```

```python
scores = df.Score
scores
```

```python
homeScore = [x[:1] for x in scores]
homeScore
awayScore = [x[-1:] for x in scores]
awayScore
```

```python
homeScore = [int(x) for x in homeScore]
awayScore = [int(x) for x in awayScore]
awayScore
```

```python
games = pd.DataFrame({"Home Team":df["Home Team"], "Home Score":homeScore,  "Away Score":awayScore,"Away Team":df["Away Team"]})
games
```

```python
manchesterDerby = games.loc[np.logical_and((games["Home Team"] == 'Manchester Utd') | (games["Home Team"] == 'Manchester City'), (games["Away Team"] == 'Manchester City') | (games["Away Team"] == 'Manchester Utd'))]
manchesterDerby
```
